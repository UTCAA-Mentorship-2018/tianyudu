```python
"""
GBM classifier.
"""
%load_ext autotime
import os
import matplotlib
import matplotlib.pyplot as plt
import sklearn
from sklearn import metrics
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble.partial_dependence import plot_partial_dependence
from sklearn.ensemble.partial_dependence import partial_dependence
from mpl_toolkits.mplot3d import Axes3D

from constants import *
from core.data.data_proc import *
from core.tools.roc_visualize import *
from ui_control import *
```

```python
DROP_THRESHOLD = 0.1
DROP_COLUMNS = []

FILE_DIR = choose_dataset(SAVED_FILE_DIRS)

df = load_data(
    file_dir=FILE_DIR,
    drop_threshold=DROP_THRESHOLD,
    drop_columns=DROP_COLUMNS)

df.drop(columns=["SK_ID_CURR"], inplace=True)

X, y = df.drop(columns=["TARGET"]), df["TARGET"]

num_fea = df.shape[1] - 1
```

```python
e, encoders = int_encode_data(X)
```

```python
# Spliting dataset.
X_train, X_test, y_train, y_test = train_test_split(
    e,
    y,
    test_size=0.2,
    shuffle=True
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train,
    y_train,
    test_size=0.25,
    shuffle=True
)
```

```python
# Scaling Data
X_scaler = StandardScaler()
X_scaler.fit(X_train.astype(np.float64))

scaled_X_train = X_scaler.transform(X_train.astype(np.float64))
scaled_X_test = X_scaler.transform(X_test.astype(np.float64))
scaled_X_val = X_scaler.transform(X_val.astype(np.float64))
```

```python
print(f"Training set: {X_train.shape}, {y_train.shape}\
\nTesting set: {X_test.shape}, {y_test.shape}\
\nValidation set: {X_val.shape}, {y_val.shape}")
```

```python
clf = GradientBoostingClassifier(
    n_estimators=500,
    verbose=1
)

clf.fit(
    X_scaler.transform(X_train),
    y_train
)

y_pred = clf.predict_proba(
    X_scaler.transform(X_test)
)
```

```python
matplotlib_roc(
    actual=y_test,
    pred_prob=y_pred[:, 1],
    show=True,
    file_dir=None
)
```

```python
# Prediction Stage
pred_train = clf.staged_predict_proba(scaled_X_train)
pred_val = clf.staged_predict_proba(scaled_X_val)
```

```python
train_hist = np.zeros((clf.n_estimators, ), dtype=np.float64)
val_hist = np.zeros((clf.n_estimators, ), dtype=np.float64)

for e, p in enumerate(pred_train):
    train_hist[e] = clf.loss_(p[:, 1], y_train)

for e, p in enumerate(pred_val):
    val_hist[e] = clf.loss_(p[:, 1], y_val)
  
```

```python
plt.plot(train_hist)
plt.plot(val_hist)
plt.legend(["Training Loss", "Validation Loss"])
plt.show()
```

```python
# PDP
names = list(X_train.columns)
names[0]
```

```python
plt.figure(figsize=(45,45))
for tar in range(55):
    plot_partial_dependence(
        clf,
        X_train,
        [tar],
        feature_names=names
    )
    plt.show()
```
